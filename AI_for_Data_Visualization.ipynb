#AI for Data Visualization Assignment

Q1. What is AI in Data Visualization? Explain how it differs from traditional data analysis.

Ans: AI in Data Visualization means using Artificial Intelligence techniques such as Machine Learning and Natural Language Processing to analyze data, identify patterns, and automatically create insights through charts, dashboards, and reports.

Traditional data analysis mainly depends on manual work. Analysts must clean data, create charts, and find trends themselves. In AI-based visualization, the system automatically detects patterns, predicts outcomes, and suggests the best visualizations.

For example, in traditional analysis a sales manager manually checks monthly sales reports. In AI visualization, Power BI or Tableau can automatically highlight that sales dropped because of delayed deliveries in a specific region.


Q2. Explain how AI helps in automatic insight generation in BI tools

Ans: AI helps BI tools generate insights automatically by analyzing large amounts of data quickly and identifying hidden patterns, trends, and anomalies.

AI can:

- Detect sudden increases or decreases in sales
- Predict future trends
- Identify factors affecting business performance
- Generate summaries in simple language.


Q3. What are the key AI features available in Microsoft Power BI? Explain any two with examples.

Ans:Some important AI features in Microsoft Power BI are:

- Key Influencers
- Smart Narratives
- Q&A Visual
- AI Insights
- Decomposition Tree
- Forecasting

1.Key Influencers

This feature identifies factors that affect a business metric.

Example:
If customer satisfaction decreases, Key Influencers may show that delivery delay and poor service quality are the main reasons.

2.Smart Narratives

This feature automatically generates written summaries of dashboard data.

Example:
Power BI may generate a sentence like:
“Sales increased by 15% this month mainly due to higher online orders.”


Q4. Explain the concept of Key Influencers and how AI identifies factors affecting a metric

Ans:Key Influencers is an AI feature in Power BI that analyzes data to determine which factors have the biggest impact on a selected metric.

AI uses statistical analysis and machine learning algorithms to compare different variables and find relationships.

For example:
If a company wants to know why employee attrition is high, AI may identify:

- Low salary
- Long working hours
- Lack of promotions

The tool displays these factors visually so decision-makers can easily understand the causes.


Q5. What is the role of Smart Narratives in AI-powered dashboards?

Ans: Smart Narratives automatically convert dashboard data into written explanations using AI.

Its main roles are:

- Simplifying complex data
- Highlighting important insights
- Saving report preparation time
- Improving communication

For example, instead of manually writing a report, Smart Narratives may automatically generate:
“North region sales grew faster than other regions due to increased festival demand.”

This helps managers quickly understand business performance.

Q6. Explain how Tableau Pulse provides proactive insights in Tableau?

Ans: Tableau Pulse is an AI-powered feature in Tableau that delivers automatic and proactive insights to users.

It continuously monitors business metrics and notifies users about important changes, trends, and anomalies.

Features of Tableau Pulse:

- Real-time monitoring
- Personalized metric tracking
- Automatic explanations
- Notifications and alerts.


Q7. What is meant by AI-driven decision-making? How does AI support business decisions?

Ans: AI-driven decision-making means using AI technologies to analyze data and support better business decisions.

AI supports decisions by:

- Predicting future outcomes
- Identifying risks and opportunities
- Providing real-time insights
- Automating analysis


Q8. Explain the importance of interpreting AI-generated insights instead of directly trusting them.

Ans: AI-generated insights should be interpreted carefully because AI systems may sometimes produce incorrect or biased results.

Reasons why interpretation is important:

- Data may contain errors
- AI may misunderstand patterns
- Business context is necessary
- Human judgment improves decision quality


Q9. Discuss the limitations of AI in data visualization tools.

Ans: Although AI improves data analysis, it also has some limitations:

1.Dependence on Data Quality
AI gives poor results if the data is incomplete or incorrect.

2.High Cost
Advanced AI tools can be expensive for small businesses.

3.Lack of Human Understanding
AI cannot fully understand emotions, culture, or business situations.

4.Bias in Results
Biased training data can produce unfair insights.

5.Complexity
Some AI models are difficult for users to understand.
